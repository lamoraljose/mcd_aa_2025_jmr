```python
import pandas as pd
import numpy as np
import sklearn
```

```python
from sklearn.datasets import fetch_california_housing
```

```python
california = fetch_california_housing()
```

```python
df = pd.DataFrame(california.data, columns=california.feature_names)
```

```python
df.head()
```

```python
# Agregamos una variable m√°s en este caso la objetivo/target
df['MedHouseValue'] = california.target
```

```python
df.head(4)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Convertimos/especificamos a float32 para optimizar memoria
df = df.astype('float32')
```

```python
df.info()
```

```python
df_indexado = df.set_index(pd.Index(range(1, len(df)+1)))
df_indexado.index.name = 'ID_Vivienda'              
```

```python
df_indexado.head(4)
```

```python
# Creamos una columna
df_indexado['Nivel_Ingreso'] = pd.cut(
    df_indexado['MedInc'],
    bins=[0, 2, 4, 6, 8, 16],
    labels=['Muy bajo', 'Bajo', 'Medio', 'Alto', 'Muy alto']
)
```

```python
df_indexado.head(4)
```

```python
# Eliminaci√≥n de columnas
df_indexado = df_indexado.drop(columns=['AveBedrms'])
```

```python
df_indexado.head(4)
```

```python
# Filtros
df_filtrado = df_indexado[df_indexado['Nivel_Ingreso'].isin(['Muy Bajo', 'Bajo'])]
```

```python
df_filtrado.head(4)
```

```python
# Filtros
df_filtrado_2 = df_indexado[(df_indexado['AveRooms'] < 3) & (df_indexado['Population'] > 500)]
```

```python
df_filtrado_2.head(4)
```

```python
# Crear subconjutos de datos
df_subconjunto = df_indexado[['MedInc', 'HouseAge', 'AveRooms', 'Population', 'MedHouseValue']]
```

```python
df_subconjunto.head(4)
```

```python
df_ingresos_altos = df_indexado[df_indexado['Nivel_Ingreso'].isin(['Medio', 'Alto', 'Muy Alto'])]
```

```python
df_ingresos_altos.head(4)
```

```python
# Grupos
df_grupo_ingreso = df_indexado.groupby('Nivel_Ingreso')['MedHouseValue'].mean().reset_index()
```
